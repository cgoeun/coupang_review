```python
# 단어는 중복을 제거하고 가능하다면 형태소분석기를 통해서도 숫자를 줄일 수 있을 수도 있다 그냥 이렇게 대중적인 활용어로 쓸 수도 있다

negative_point = 0.0
positive_point = 0.0

for idx, word in enumerate(word_list) :
    for i in range(len(data)): 
        if re.search(word, data['review_r'].iloc[i]) or re.search(word, data['review'].iloc[i]):
            print(i)
            if data.iloc[i]['star'] == 5 : 
                positive_point += 4/101203*1000000
            elif data.iloc[i]['star'] == 4 :
                positive_point += 3/21465*1000000 
                negative_point += 1/21465*1000000 
            elif data.iloc[i]['star'] == 3 :
                positive_point += 2/8661*1000000 
                negative_point += 2/8661*1000000 
            elif data.iloc[i]['star'] == 2 :
                positive_point += 1/3809*1000000 
                negative_point += 3/3809*1000000 
            elif data.iloc[i]['star'] == 1 :
                negative_point += 4/9744*1000000   
                
    dictionary_df.iloc[idx,1] = positive_point
    dictionary_df.iloc[idx,2] = negative_point
    negative_point = 0.0
    positive_point = 0.0

    
    # print(positive_point-negative_point) # 차이가 +이면 positive 값  -이면 negative 값 이 값이 해당 단어의 사전값(딕셔너리값)이 될 것 같다.
                                         # 차이가 별로 안 나는 값은 최종적으로 사용할 사전에서 제거한다.

    # 전체 라벨링은 단어사전(딕셔너리)을 통해 단어값들을 다 더해서 +일 때 긍정, -일 때 부정으로 처리해보도록 하겠다.
                
        
```
