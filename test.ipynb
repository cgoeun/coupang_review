```python
import re

def find_not_text(text):
    pattern = '[\w\s]'         # 특수기호제거
    not_text = re.sub(pattern=pattern, repl='', string=text)
    not_text = re.sub('[.,?\(\)\:\\\]', '', not_text)
    # not_text = re.sub('[-=+#/\:^$@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', repl='', string=text)
    return not_text   

# len(find_not_text(list(data.iloc[1])[3]))
```

```python
def find_text(text):
    pattern = '[^\w\sㄱ-ㅎㅏ-ㅣ]'         # 특수기호제거
    text = re.sub(pattern=pattern, repl='', string=text)
    text = re.sub(pattern='[\s]+', repl=' ', string=text)
    # not_text = re.sub('[-=+#/\:^$@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', repl='', string=text)
    return text   
```
